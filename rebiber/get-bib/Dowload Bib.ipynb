```python
import requests
```

```python
confs=['ecml', 'wacv']#['eccv', 'iccv', 'bmvc','cvpr', 'accv', 'neurips','miccai', 'ecml']
years=list(range(2000, 2024))
```

```python
years
```

```python
for conf in confs : 
    for year in years : 
        cites = ''
        for step in range(5) :
            print(step, conf, year)
            s=f'https://dblp.org/search/publ/api?q=conf/{conf}/{year}&h=1000&f={step*1000}&format=bib'
            cs = requests.get(s).text
            if cs == '' : 
                print('stop')
                break
            cites+= cs
        with open(f'/Users/etiennemeunier/Desktop/rebiber/rebiber/raw_data/{conf}{year}.bib', 'w') as f : 
            f.write(cites)

        with open(f'bibs/{conf}{year}.bib', 'w') as f : 
            f.write(cites)
```

```python
for conf in confs : 
    for year in years : 
        !(cd /Users/etiennemeunier/Desktop/rebiber/rebiber/;./add_conf.sh {conf} {year})
```
